```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from typing import *
```

```python
def load_acc_rewards(agent_name: str, filename: str) -> Tuple[List[int], int]:
    acc_rewards = []
    with open(f"agent_code/{agent_name}/models/{filename}") as fp:
        line= fp.readline()
        while line:
            acc_rewards.append(int(line.strip()))
            line = fp.readline()
    
    return acc_rewards, len(acc_rewards)

def calc_rewards_per_episode(acc_rewards: List[int]) -> List[int]:
    rewards= []
    for idx, acc_r in enumerate(acc_rewards):
        if idx == 0:
            rewards.append(acc_rewards[idx])
        else:
            rewards.append(acc_rewards[idx] - acc_rewards[idx-1])
    return rewards
```

```python
acc_rewards_double, n_rounds = load_acc_rewards("task1_double_q", "rewards-task1_double_q-03-12T22:50.csv")
rewards_double = calc_rewards_per_episode(acc_rewards_double)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
ax1.plot(np.arange(0, n_rounds),acc_rewards_double)
ax1.set_title('Accumulated rewards')
ax2.plot(np.arange(0, n_rounds),rewards_double)
ax2.set_title("Rewards per Episode")
```
